```python
import requests, nltk
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

# Define the stock symbol and number of messages to retrieve
symbol = 'AAPL'
limit = 100

# Fetch StockTwits messages for the stock symbol
url = f"https://api.stocktwits.com/api/2/streams/symbol/{symbol}.json?limit={limit}"
response = requests.get(url)
data = response.json()

# Extract messages from the response
messages = [msg['body'] for msg in data['messages']]

# Initialize the Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()

# Perform sentiment analysis on each message
sentiments = []
for message in messages:
    sentiment_score = sia.polarity_scores(message)
    sentiment = 'Positive' if sentiment_score['compound'] >= 0 else 'Negative'
    sentiments.append(sentiment)

# Count the number of positive and negative sentiments
positive_count = sentiments.count('Positive')
negative_count = sentiments.count('Negative')

# Print the sentiment analysis results
print("Sentiment Analysis Results")
print("----------------------------")
print(f"Total Messages: {len(messages)}")
print(f"Positive Sentiments: {positive_count}")
print(f"Negative Sentiments: {negative_count}")
```

```python
import requests
from nltk.sentiment import SentimentIntensityAnalyzer

# Define the stock symbol and number of messages to retrieve
symbol = 'AAPL'
limit = 100
total_messages = []

# Function to fetch messages using pagination
def fetch_messages(symbol, limit, max_id=None):
    url = f"https://api.stocktwits.com/api/2/streams/symbol/{symbol}.json?limit={limit}&max={max_id}"
    response = requests.get(url)
    data = response.json()
    messages = data['messages']
    total_messages.extend([msg['body'] for msg in messages])
    cursor = data['cursor']
    if cursor['more']:
        max_id = cursor['max']
        fetch_messages(symbol, limit, max_id)

# Fetch StockTwits messages for the stock symbol
fetch_messages(symbol, limit)

# Initialize the Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()

# Perform sentiment analysis on each message
sentiments = []
for message in total_messages:
    sentiment_score = sia.polarity_scores(message)
    sentiment = 'Positive' if sentiment_score['compound'] >= 0 else 'Negative'
    sentiments.append(sentiment)

# Count the number of positive and negative sentiments
positive_count = sentiments.count('Positive')
negative_count = sentiments.count('Negative')

# Print the sentiment analysis results
print("Sentiment Analysis Results")
print("----------------------------")
print(f"Total Messages: {len(total_messages)}")
print(f"Positive Sentiments: {positive_count}")
print(f"Negative Sentiments: {negative_count}")
```

```python
import requests
from textblob import TextBlob

# Define the stock symbol and number of messages to retrieve
symbol = 'AAPL'
limit = 100
total_messages = []

# Fetch StockTwits messages for the stock symbol
url = f"https://api.stocktwits.com/api/2/streams/symbol/{symbol}.json?limit={limit}"
response = requests.get(url)
data = response.json()
messages = data['messages']
total_messages.extend([msg['body'] for msg in messages])

# Perform sentiment analysis on each message
sentiments = []
for message in total_messages:
    sentiment = TextBlob(message).sentiment.polarity
    if sentiment > 0:
        sentiments.append('Positive')
    elif sentiment < 0:
        sentiments.append('Negative')
    else:
        sentiments.append('Neutral')

# Count the number of positive, negative, and neutral sentiments
positive_count = sentiments.count('Positive')
negative_count = sentiments.count('Negative')
neutral_count = sentiments.count('Neutral')

# Print the sentiment analysis results
print("Sentiment Analysis Results")
print("----------------------------")
print(f"Total Messages: {len(total_messages)}")
print(f"Positive Sentiments: {positive_count}")
print(f"Negative Sentiments: {negative_count}")
print(f"Neutral Sentiments: {neutral_count}")
```

```python

import nest_asyncio
nest_asyncio.apply()

import twint
from nltk.sentiment import SentimentIntensityAnalyzer

# Define the stock symbol
symbol = 'AAPL'

# Configure the Twint search parameters
c = twint.Config()
c.Search = f"${symbol}"
c.Limit = 100  # Number of tweets to retrieve
c.Pandas = True  # Store results in a Pandas DataFrame

# Run the Twint search
twint.run.Search(c)

# Get the tweets as a Pandas DataFrame
tweets_df = twint.storage.panda.Tweets_df

# Extract the tweet text
tweets = tweets_df['tweet'].tolist()

# Initialize the Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()

# Perform sentiment analysis on each tweet
sentiments = []
for tweet in tweets:
    sentiment_score = sia.polarity_scores(tweet)
    sentiment = 'Positive' if sentiment_score['compound'] >= 0 else 'Negative'
    sentiments.append(sentiment)

# Count the number of positive and negative sentiments
positive_count = sentiments.count('Positive')
negative_count = sentiments.count('Negative')

# Print the sentiment analysis results
print("Sentiment Analysis Results")
print("----------------------------")
print(f"Total Tweets: {len(tweets)}")
print(f"Positive Sentiments: {positive_count}")
print(f"Negative Sentiments: {negative_count}")
```
